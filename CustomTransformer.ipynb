```python
"""

FunctionTransformer 

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.FunctionTransformer.html

applies a custom function to columns in your dataframe

1. Execute the code 
   (in Jupyter, split it into multiple cells)

2. Understand what is happening

3. Explain to the rest of the group what you did
"""

import pandas as pd
from sklearn.preprocessing import FunctionTransformer

df = pd.read_csv('all_penguins_clean.csv')
df.head()
```

```python
egg = df['Date Egg']
```

```python
year = [int('20' + x.split('/')[-1]) for x in egg]
```

```python
def get_year(eggDate):
    list_year = []
    
    for x in eggDate:
        list_year.append(int('20' + x.split('/')[-1]))
    return list_year
    
```

```python
yearer= FunctionTransformer(get_year)
cols = df['Date Egg']
yearer.fit(cols)
df['EggYear'] = yearer.transform(cols)
```
